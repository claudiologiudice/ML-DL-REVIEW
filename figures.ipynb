```r
f = seq(-5,4,0.001)
loss01 = function(z){return(ifelse(z<=0,1,0))}
perceptron = function(z){return(sapply(z,function(x){return(max(0,-x));}))}
hinge = function(z){return(sapply(z, function(x){return(max(0, 1-x))}))}
logloss = function(z){return(sapply(z, function(x){return((1/log(2))*log(1+exp(-2*x)))}))}
sigmoidloss = function(z){return(sapply(z, function(x){return(1-1/(1+exp(-x)))}))}

colors=c("black","blue","red","cyan")
linetype=c(1,2,3,4)

#pdf("losses.pdf", width=5, height=5, pointsize=10)
plot(f, loss01(f), type='l', ylim=c(-1,4), cex.lab=1, col=colors[1], lty=1, xlab="z=yÂ·f(x)", ylab="Loss size")
points(f, perceptron(f), type='l', col=colors[2], lty=2)
points(f, hinge(f), type='l', col=colors[3], lty=3)
#points(f, sigmoidloss(f), type='l', col=colors[4], lty=linetype[4])
points(f, logloss(f), type='l', col=colors[4], lty=4)
#points(f, -tanh(f), type='l', col=colors[6], lty=linetype[6])
legend(1.5, 4, legend=c("0-1", "perceptron", "hinge", "log"),
       col=colors, lty=linetype, cex=1, bty = "n")
text(-2.7, -1, "incorrectly classified yf(x)<0", cex=1, col="dark gray")
text(2.2, -1, "correctly classified yf(x)>=0", cex=1, col="dark gray")
#segments(1,-1.5,1,-0.7, lty=2, col="dark gray", lwd=2)
segments(1,-0.4,1,4, lty=2, col="dark gray")
text(0.5, -0.3, "yf(x)<1", cex=0.9, col="red")
segments(0,-1.5,0,0, lty=2, col="dark gray")
dev.off()


```
